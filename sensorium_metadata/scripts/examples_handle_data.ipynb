The data can be access using the method get_data

```python
# data not normalized
data = response.get_data() 
print(data.shape)

# data normalized by the mean (across the responses for all trials) (d - mean)/mean
data_norm_mean = response.get_data(normalization='by_mean') 

# data normalized by the standard deviation (across the responses for all trials) d/std
data_norm_std = response.get_data(normalization='by_std')
```

```python
from managedata.handle_dataset import combine_data
from managedata.behavioral import compute_power_spectrum
```

```python
pupil_data_avg = combine_data(pupil)

pupil_detrended = [p.copy(deep=True).detrend() for p in pupil]
pupil_data_detrended_avg = combine_data(pupil_detrended)

fig, ax = plt.subplots(ncols=1,nrows=1)
ax.plot(pupil_data_avg)

fig, ax = plt.subplots(ncols=1,nrows=1)
ax.plot(pupil_data_detrended_avg)
```

```python
rec = 'dynamic29514-2-9-Video-8744edeac3b4d1ce16b680916b5267ce'
lab = 'GaussianDot'
pupil, triasl_df = dat.load_behavior_by('pupil', recording=rec, label=lab)

psd = []
for p in pupil:
    freqs, psd_i = p.compute_power_spectrum()
    psd.append(psd_i)
psd = np.array(psd)
psd_mean = np.mean(psd, axis=0)

fig, ax = plt.subplots(ncols=1,nrows=1)
ax.semilogy(freqs, psd.T)
ax.set_xlim(0, 2)
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("PSD")
ax.set_title("Pupil size power spectrum")
plt.show()

fig, ax = plt.subplots(ncols=1,nrows=1)
ax.semilogy(freqs, psd_mean)
ax.set_xlim(0, 2)
ax.set_xlabel("Frequency (Hz)")
ax.set_ylabel("PSD")
ax.set_title("Pupil size power spectrum")
plt.show()
```
