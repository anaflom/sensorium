```python
import json
import os
from pathlib import Path
from tqdm import tqdm

```

```python
def insert_after(d, after_key, new_key, new_value):
    items = []
    for k, v in d.items():
        items.append((k, v))
        if k == after_key:
            items.append((new_key, new_value))
    return dict(items)
```

# Add sampling frequency to the metadata generated for each video trial

```python
folder_meta = '/home/anaflo/MDMC/thesis/sensorium/sensorium_metadata/metadata'

# results folder
folder_results = os.path.join('..','intermediate_results')

all_recording_folders = [
    'dynamic29156-11-10-Video-8744edeac3b4d1ce16b680916b5267ce',
    'dynamic29228-2-10-Video-8744edeac3b4d1ce16b680916b5267ce',
    'dynamic29234-6-9-Video-8744edeac3b4d1ce16b680916b5267ce',
    'dynamic29513-3-5-Video-8744edeac3b4d1ce16b680916b5267ce',
    'dynamic29514-2-9-Video-8744edeac3b4d1ce16b680916b5267ce',
]
```

```python
for recording in all_recording_folders:
    folder_metadata_videos = os.path.join(folder_results,recording, 'meta_videos')
    folder = Path(folder_metadata_videos)
    json_files = list(folder.glob("*.json"))

    for file in tqdm(json_files):       
        # load
        with open(file, "r", encoding="utf-8") as f:
            metavideo = json.load(f)

        if 'sampling_freq' not in metavideo.keys():
            # add sampling frequnecy
            metavideo_new = insert_after(metavideo, "ID", "sampling_freq", 30)
            # save
            with open(file, "w") as f:
                json.dump(metavideo_new, f, indent=4)
```
