# Data Loading example

This notebooks shows how to load the neural responses for all trials per recording

```python
from pathlib import Path

from ssdatam.dataset import DataSet
```

Initialize a `DataSet` object to handle the full dataset.

```python
repo_root = Path.cwd().parent

# path to the folder with the data as downloaded
folder_data = repo_root / 'data/'

# path to the metadata folder
folder_meta = repo_root / 'metadata'

# initialize an object to handle the dataset
ds = DataSet(folder_data, folder_metadata=folder_meta)

# load trials metadata
trials_df = ds.get_trials_metadata()
```

Loop over recordings, labels, and video IDs and load the responses 

```python

labels = ['NaturalVideo','NaturalImages','GaussianDot','Gabor','PinkNoise','RandomDots']

# loop over recordings (mice)
for rec in ds.recording:
    # loop over labels
    for lab in labels:
        trials_df = ds.filter_trials(recording=rec, label=lab)
        if len(trials_df)>0:
            # loop over videos IDs
            ids_list = set(trials_df['ID'].to_list())
            for id in ids_list:
                responses, _ = ds.load_responses_by(recording=rec, label=lab, ID=id, valid_trial=True)
                print(f"Recording {rec} - Label {lab} - ID {id}: {len(responses)} trials found")
                # loop over trials with same video ID
                for response in responses:
                    # get the data normalized by the mean (across the responses for all trials) (d - mean)/mean
                    data_norm_mean = response.get_data(normalization='by_mean') 
                    # get the neurons coordinates
                    coord_xyz = response.neurons.coord_xyz
```
